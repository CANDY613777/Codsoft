```python
#Task 1: To-do list

def add():
    task=input("Enter the task to be added: ")
    to_do_list.append(task);
    print("Your task has been added to your to do list! All the best :)")
def update():
    n=int(input("Which task is to be updated? Enter position: "))
    task=input("Enter new task description: ")
    if(n<1 or n>len(to_do_list)):
        print("Invalid input!!!")
    else:
        to_do_list[n-1]=task
def complete():
    n=int(input("Which task has been completed? Enter position: "))
    if(n<1 or n>len(to_do_list)):
        print("Invalid input!!!")
    else:
        to_do_list[n-1]=to_do_list[n-1]+": completed"
        print("Congratulations on completing your task successfully! Keep it up!")
def to_do_view():  
    for i in to_do_list:
        print(i)
b=True
while (b==True):
    print("Hello and welcome!\nChoose an option:\n1. Create a to-do list\n2. Add tasks\n3. Update to-do-list \n4. Mark tasks as complete \n5. View your progress\n6. Exit")
    n=int(input("Enter your choice: "))
    if(n==1):
        to_do_list=[]
        completed_list=[]
        print("Your to-do list has been created successfully!")
    elif(n==2):
        add()
    elif(n==3):
        update()
    elif(n==4):
        complete()
    elif(n==5):
        to_do_view()
    else:
        b=False
```

```python
#Task 4: Rock-paper-scissors
import random
s1=0
s2=0
while(True):
    choices=["rock", "paper", "scissors"]
    ch=input("Enter your choice: rock, paper, scissors.\nPress "F" to exit: ")
    ch=ch.lower()
    if(ch=="f"):
        print("Your score: "+str(s1)+"\nComputer's score: "+str(s2))
        if(s2==s1):
            print("It's a tie!")
        elif(s2>s1):
            print("Computer wins!")
        else:
            print("You win!")
        break
    else:
        ch1=random.choice(choices)
        print("Your choice: "+ch+"\nComputer's choice: "+ch1)
        if(ch1==ch):
            print("It's a tie!")
        elif((ch=="rock" and ch1=="paper")or(ch=="paper" and ch1=="scissors") or (ch=="scissors" and ch1=="rock")):
            print("Computer wins!")
            s2=s2+1
        else:
            print("You win!")
            s1=s1+1
```
